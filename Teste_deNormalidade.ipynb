```python
from scipy import stats
from scipy.stats import norm,skewnorm
import matplotlib.pyplot as plt
```

```python
dados = norm.rvs(size=1000)
dados
```

```python
plt.hist(dados,bins=20)
plt.title("DADOS")
```

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados)
#TESTE DE HIPOTESE
```
